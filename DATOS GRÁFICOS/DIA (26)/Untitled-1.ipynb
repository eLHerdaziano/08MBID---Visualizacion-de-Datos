```python
import pandas as pd
import altair as alt

# Leer los datos del archivo Excel
df = pd.read_csv("private-investment-in-artificial-intelligence.csv", delimiter=",")

# Convertir la columna 'World ($)' a millones de euros
df['World ($, million euros)'] = df['World ($)'] / 1000000

# Crear el gráfico de streamgraph
streamgraph = alt.Chart(df).mark_area(interpolate='basis').encode(
    x=alt.X('Year:O', title='Año'),
    y=alt.Y('sum(World ($, million euros)):Q', stack="center", axis=alt.Axis(title='Inversión (millones de €)', grid=True)),
    color=alt.Color('Entity:N', title='Categoría', scale=alt.Scale(scheme='category10'))
).properties(
    width=800,
    height=400,
    title=alt.TitleParams(text='Inversiones privadas en IA', subtitle='Gasto en euros en sectores privados en el mundo'),
).configure_axis(
    labelFontSize=12,    # Tamaño de fuente de las etiquetas de los ejes
    titleFontSize=14,    # Tamaño de fuente de los títulos de los ejes
    gridOpacity=0.3      # Opacidad de las líneas de la cuadrícula
).configure_view(
    stroke=None
)

# Mostrar el gráfico
streamgraph.show()
```
