```python
import pandas as pd
import matplotlib.pyplot as plt
import joypy

# Leer los datos del archivo Excel
df = pd.read_excel("Nuevo Hoja de cálculo de Microsoft Excel.xlsx")

# Contar la frecuencia de cada categoría en 2017
top_categories_2017 = df['Servicio'].value_counts().nlargest(10).index.tolist()

# Filtrar el DataFrame para incluir solo las 10 categorías más comunes en 2017
df_filtered_2017 = df[df['Servicio'].isin(top_categories_2017)]

# Calcular el valor medio de satisfacción para cada categoría en 2017
mean_satisfaction_2017 = df_filtered_2017.groupby('Servicio')['Encuesta 2017'].mean()

# Crear el ridgeline plot para 2017
plt.figure(figsize=(10, 6))
joypy.joyplot(df_filtered_2017, by="Servicio", column="Encuesta 2017", figsize=(12,8), alpha=0.6, ylabelsize=8)

plt.suptitle('Sondeo de satisfacción de los servicios del Ayuntamiento de Zaragoza - 2017',y=1.05)
plt.xlabel('Satisfacción')
plt.ylabel('Servicio')

# Agregar una cuadrícula
plt.grid(True)

plt.show()
```
