```python
import pandas as pd
import matplotlib.pyplot as plt
import joypy

# Leer los datos del archivo Excel
df = pd.read_excel("d89af88a-d055-481b-aa31-05e9e35061ef.csv")

# Contar la frecuencia de cada categor√≠a en 2017
top_categories_2017 = df['Servicio'].value_counts().nlargest(10).index.tolist()

# Filtrar el DataFrame para incluir solo las 10 categor√≠as m√°s comunes en 2017
df_filtered_2017 = df[df['Servicio'].isin(top_categories_2017)]

# Calcular el valor medio de satisfacci√≥n para cada categor√≠a en 2017
mean_satisfaction_2017 = df_filtered_2017.groupby('Servicio')['Encuesta 2017'].mean()

plt.style.use('dark_background')  # Fondo oscuro
plt.rc('font', family='sans-serif', size=12)  # Fuente y tama√±o de la letra

# Crear el ridgeline plot para 2017
fig, axes = joypy.joyplot(df_filtered_2017, by="Servicio", column="Encuesta 2017", figsize=(8, 8), alpha=0.75, 
                           xlabelsize=8, ylabelsize=8, overlap=0.7,bw_method=0.60)  # Ajusta este valor seg√∫n tus datos

# T√≠tulo y etiquetas
plt.suptitle('Sondeo de satisfacci√≥n de los servicios del Ayuntamiento de Zaragoza - 2017', y=1.02, fontsize=14)
plt.xlabel('Satisfacci√≥n', fontsize=10)
plt.ylabel('Servicio', fontsize=10)

# Ajustar los ejes
for ax in axes:
    ax.set_yticklabels(ax.get_yticklabels(), fontsize=7)  # Tama√±o de las etiquetas del eje y

# Agregar una cuadr√≠cula
plt.grid(True, alpha=0.5)


plt.show()
```
