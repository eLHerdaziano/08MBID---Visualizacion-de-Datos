```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.image as mpimg
from PIL import Image


df = pd.read_csv('AAPL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_resampled = df.resample('M', on='Date').mean() 

plt.figure(figsize=(16, 6))
fechas = [pd.Timestamp('1998-01-31'),pd.Timestamp('1998-07-31'), pd.Timestamp('2001-01-31'), pd.Timestamp('2008-01-31'), pd.Timestamp('2008-07-31')]
colores = ['#ADD8E6', '#000000', '#A9A9A9', '#F5DEB3','#444444' ]  # Colores proporcionados en formato #RGB
label = ['Cambio del logo rainbow al logo azul claro','Cambio al logo negro','Cambio al logo plata','Cambio al logo blanco','Cambio al logo actual',]

i = [0.2,0.4,0.6,0.8]
sns.lineplot(y=df_resampled['Close'], x=df_resampled.index)
for fecha, color,etiqueta in zip(fechas, colores,label):
    plt.axvline(fecha, color=color, linestyle='--', label=f'{etiqueta} ({fecha.year})')
plt.legend()

plt.title('Valor en dólares del cierre de mercado junto con el cambio de imagen del logo de la marca. Fuente: KAGGLE')
plt.ylabel('Valor ($)')
plt.xlabel('Año')
plt.suptitle('Influencia del cambio de imagen en el valor de las acciones de Apple',fontsize=18,fontweight='bold')
image = Image.open('LOGOTIPO-APPLE-evolucion1.png')
print(f"Original size : {image.size}") #

sunset_resized = image.resize((500, 120))

plt.figimage(sunset_resized, xo=80, yo=250, alpha=1)
plt.text(0.5, -0.15, 'Fuente: Kaggle Nota: No se tiene en cuenta el cambio del logo de Newton dado que aún no había entrado en bolsa', ha='center', va='center', fontsize=10, transform=plt.gca().transAxes)

plt.show()
```
