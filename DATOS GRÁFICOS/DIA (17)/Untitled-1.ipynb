```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import networkx as nx

# Leer los datos del archivo CSV
df = pd.read_csv('evolucion_de_los_votos_que_espa√±a_ha_recibido_en_eurovision.csv', encoding='latin-1')

# Crear un gr√°fico dirigido desde el DataFrame
G = nx.from_pandas_edgelist(df, '√Ø¬ª¬øFrom', 'To', create_using=nx.DiGraph())

# Obtener el n√∫mero de votos que recibe cada pa√≠s
vote_counts = df['To'].value_counts().to_dict()

# Crear un mapeo de tama√±o de nodo basado en el n√∫mero de votos
min_node_size = 100  # Tama√±o m√≠nimo para los nodos sin votos
node_size = [max(vote_counts.get(node, 0) * 100, min_node_size) for node in G.nodes()]

# Dibujar el gr√°fico de red con disposici√≥n Fruchterman-Reingold y mayor dispersi√≥n
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.2)  # Usar disposici√≥n Fruchterman-Reingold con mayor dispersi√≥n (k m√°s peque√±o)
nx.draw(G, pos, with_labels=True, node_size=node_size, node_color='skyblue', edge_color='gray', font_size=8, arrowsize=12)

# A√±adir etiquetas de pa√≠s
pos_labels = {key: (x, y + 0.05) for key, (x, y) in pos.items()}  # Mover las etiquetas hacia arriba

plt.title('Pa√≠ses que recibieron 12 puntos en la votaci√≥n de Eurovisi√≥n 2023')
plt.axis('off')  # Ocultar ejes
plt.tight_layout()  # Ajustar el dise√±o
plt.show()
```
