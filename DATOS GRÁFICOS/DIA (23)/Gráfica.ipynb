```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
df = pd.read_csv("SN_m_tot_V2.0.csv", encoding='latin1')


# Definir la misma escala para ambos heatmaps
vmin = min(df['Valor'])
vmax = max(df['Valor'])

# Heatmap para la temperatura mxima
plt.figure(figsize=(16,4))
sns.heatmap(df.pivot(index='Mes',columns='Ao',values='Valor'))
plt.title('La actividad solar flucta en ciclos aproximadamente cada 11 aos, manifestndose en cambios en la cantidad de manchas solares y eventos como llamaradas y eyecciones de masa coronal.\n Estos ciclos son fundamentales para comprender el comportamiento del Sol y su influencia en el clima y la tecnologa espacial.')
plt.xlabel('Ao')
plt.ylabel('Mes del ao')



# Agregar ttulo y subttulo
plt.suptitle('Actividad del sol a lo largo de los ltimos 275 aos', fontsize=16, y=1,x=0.4)


plt.tight_layout()
plt.show()
```
