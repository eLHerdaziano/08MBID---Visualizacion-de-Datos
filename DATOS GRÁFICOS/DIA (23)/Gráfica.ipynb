```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
df = pd.read_csv("SN_m_tot_V2.0.csv", encoding='latin1')


# Definir la misma escala para ambos heatmaps
vmin = min(df['Valor'])
vmax = max(df['Valor'])

# Heatmap para la temperatura máxima
plt.figure(figsize=(16,4))
sns.heatmap(df.pivot(index='Mes',columns='Año',values='Valor'))
plt.title('La actividad solar fluctúa en ciclos aproximadamente cada 11 años, manifestándose en cambios en la cantidad de manchas solares y eventos como llamaradas y eyecciones de masa coronal.\n Estos ciclos son fundamentales para comprender el comportamiento del Sol y su influencia en el clima y la tecnología espacial.')
plt.xlabel('Año')
plt.ylabel('Mes del año')



# Agregar título y subtítulo
plt.suptitle('Actividad del sol a lo largo de los últimos 275 años', fontsize=16, y=1,x=0.4)


plt.tight_layout()
plt.show()
```
