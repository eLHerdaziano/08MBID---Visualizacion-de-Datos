```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer los datos del archivo Excel
datos = pd.read_excel('2_3_NI_Sex_Edad(1)_Amb_CA.xlsx', index_col=0)

# Filtrar las categoras de inters
categorias = [
    'Total - Hombres',
    'De 18 a 21 aos - Hombres',
    'De 22 a 25 aos - Hombres',
    'De 26 a 30 aos - Hombres',
    'Ms de 30 aos - Hombres'
]
categorias_titulos = [
    'Total',
    'De 18 a 21 aos',
    'De 22 a 25 aos',
    'De 26 a 30 aos',
    'Ms de 30 aos'
]



plt.figure(figsize=(16, 20))

for i, (categoria, categorias_encabezado) in enumerate(zip(categorias, categorias_titulos)):
    # Obtener los datos de hombres y mujeres para la categora actual
    datos_hombres = datos.loc[categoria]
    datos_mujeres = datos.loc[categoria.replace('Hombres', 'Mujeres')]

    # Calcular la media de los datos
    media_hombres = datos_hombres.mean()
    media_mujeres = datos_mujeres.mean()

    # Crear el violin plot para la categora actual
    plt.subplot(3, 2, i + 1)
    sns.violinplot(data=[datos_hombres, datos_mujeres], linewidth=1.5)
    plt.xlabel('Gnero', fontsize=12, fontweight='bold')
    plt.ylabel('Tasa de nuevos ingresos', fontsize=12, fontweight='bold')
    plt.title(f'Distribucin de la tasa de nuevos ingresos -> {categorias_encabezado}', fontsize=14, fontweight='bold')
    plt.text(0.5, 0.95, f'Media Hombres: {media_hombres:.2f}\nMedia Mujeres: {media_mujeres:.2f}',
             horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10)

plt.suptitle('Distribucin de la tasa de nuevos ingresos de estudiantes en las universidades de Espaa por categora y gnero. Aos 2008-2020', fontsize=20, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
