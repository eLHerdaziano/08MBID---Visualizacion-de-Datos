```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.image as mpimg
from PIL import Image

# Make our graphics a little prettier
plt.style.use('ggplot')
plt.figure(figsize=(8, 16))  # Aumenta el tamao verticalmente
df = pd.read_csv('evolucion_de_la_generacion_de_energias_renovables_en_espaa_por_tipo_de_renovable.csv')

df['dia'] = 1

meses_mapping = {
    'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4, 'Mayo': 5, 'Junio': 6,
    'Julio': 7, 'Agosto': 8, 'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12
}

df['Periodo'] = df['Periodo'].map(meses_mapping)

df['date_str'] = df['Ao'].astype(str) + '-' + df['Periodo'].astype(str) + '-' + df['dia'].astype(str)

df['date'] = pd.to_datetime(df['date_str'])
df['date'] = df['date'].dt.strftime('%Y-%m')
df.set_index('date')
df = df.drop(['Ao', 'Periodo', 'dia', 'date_str', 'date'], axis=1)
df = df.astype(int)

df.plot(kind='area', stacked=True)

plt.legend(loc=(1.04, 0))
plt.show()
```
